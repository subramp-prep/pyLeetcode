## 051 - N Queens

The n-queens puzzle is the problem of placing n queens on an n×n chessboard such that no two queens attack each other.




Given an integer n, return all distinct solutions to the n-queens puzzle.

Each solution contains a distinct board configuration of the n-queens' placement, where 'Q' and '.' both indicate a queen and an empty space respectively.

**Example:**
```
Input: 4
Output: [
 [".Q..",  // Solution 1
  "...Q",
  "Q...",
  "..Q."],

 ["..Q.",  // Solution 2
  "Q...",
  "...Q",
  ".Q.."]
]
```
Explanation: There exist two distinct solutions to the 4-queens puzzle as shown above.

````python
class Board:
    def isValid(self, row, column):
        #navigate same column
        for c in range(self.column):
            if self.board[row][c] == 'Q':
                return False

        # navigate 135 degrees
        for i, j in zip(range(row, -1, -1), range(column, -1, -1)):
            if self.board[i][j] == 'Q':
                return False

        # navigate 45 degrees
        for i, j in zip(range(row, self.row, 1), range(column, -1, -1)):
            if self.board[i][j] == 'Q':
                return False

        return True


    def solveNQueens(self, N):
        self.row = N
        self.column = N
        self.board = [['.' for x in range(N)] for y in range(N)]

        output = []
        self.queensHelper(0, output)

        return output

    def queensHelper(self, column, output):
        if column >= self.column:
            # list comprehension
            b = [''.join(r) for r in self.board]
            
            #for i in range(self.row):
            #    b.append(''.join(self.board[i]))
            
            output.append(b)
            return
        else:
            for row in range(self.row):
                #print("row:", row, "column:", column)
                if self.isValid(row, column):
                    #choose
                    self.board[row][column] = 'Q'

                    #explore
                    self.queensHelper(column + 1, output)
                    
                    #unchoose
                    self.board[row][column] = '.'
````

````python
b = Board()
combinations = b.solveNQueens(4)

for i, board in enumerate(combinations):
    print('Combination ', i+1)
    for line in board:
        print(line)
    print(' ')
````
